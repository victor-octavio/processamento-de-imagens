# MORFOLOGIA MATEM√ÅTICA

```python
#Bibliotecas
import cv2
from matplotlib import pyplot as plt
import numpy as np
import PIL.Image as Image
plt.style.use("dark_background")
```

# FINGER PRINT <ABERTURA E FECHAMENTO> --> ERODIR - DILATAR- DILATAR - ERODIR

```python
# IMPORTAR A IMAGEM
img2 = cv2.imread(r"Fig0911(a)(noisy_fingerprint).tif")

# NORMALIZAR A IMAGEM
img_n = (img2 - img2.min())/(img2.max())-(img2.min())

plt.figure(figsize=(10,10))
plt.imshow(img2, cmap="Greys");
```

# **1¬∞ EROS√ÉO**

```python
# KERNEL (k)
N = 3
KR= cv2.getStructuringElement(cv2.MORPH_RECT,(N,N)) #   Mascara retangular
KE= cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(N,N)) #  Mascara eliptica
KC= cv2.getStructuringElement(cv2.MORPH_CROSS,(N,N)) #   Mascara em cruz

KQ = KR # escolha da mascara

dilate = cv2.dilate(img_n, KQ, iterations=1)
erode = cv2.erode(img_n, KQ, iterations = 1)

plt.figure(figsize=(18,10));
plt.subplot(1,2,1);plt.imshow(img_n, cmap ='gray');plt.title('imagem original',size=25)
plt.subplot(1,2,2);plt.imshow(erode, cmap ='gray');plt.title('imagem erodida',size=25);
```

# **1¬∞ DILATA√á√ÉO**

```python
dilate = cv2.dilate(erode, KR)
plt.figure(figsize=(18,10));
plt.subplot(1,2,1);plt.imshow(img_n, cmap ='gray');plt.title('imagem original',size=25)
plt.subplot(1,2,2);plt.imshow(dilate, cmap ='gray');plt.title('imagem dilatada',size=25);
```

# **2¬∞ DILATA√á√ÉO**

```python
dilatacao_abertura = cv2.dilate(dilate, KR)
plt.figure(figsize=(18,10));
plt.subplot(1,2,1);plt.imshow(img_n, cmap ='gray');plt.title('imagem original',size=25)
plt.subplot(1,2,2);plt.imshow(dilatacao_abertura, cmap ='gray');plt.title('imagem dilatada abertura',size=25);
```

# **2¬∞ EROS√ÉO**

```python
erosao_fechamento = cv2.erode(dilatacao_abertura, KR)
plt.figure(figsize=(18,10));
plt.subplot(1,2,1);plt.imshow(img_n, cmap ='gray');plt.title('imagem original',size=25)
plt.subplot(1,2,2);plt.imshow(erosao_fechamento, cmap ='gray');plt.title('imagem fechamento',size=25);
```
